```python
#creating subplots of various weather conditions
import pandas as pd
df=pd.read_csv('testset.csv')
df.fillna(df.mean(),inplace=True)
df[' conds'].fillna(df[' conds'].mode(), inplace=True)
df[' _precipm'].fillna("bisht", inplace=True)
df[' _wdire'].fillna(df[" _wdire"].mode(), inplace=True)

result=df[df['datetime_utc'].str.contains('19971015')]

import matplotlib.pyplot as plt
fig1=plt.figure()
fig2=plt.figure()
fig3=plt.figure()
fig4=plt.figure()
import numpy as np
plt1=fig1.add_subplot(221)
plt2=fig1.add_subplot(222)
plt3=fig1.add_subplot(223)
plt4=fig1.add_subplot(224)
plt5=fig2.add_subplot(221)
plt6=fig2.add_subplot(222)
plt7=fig2.add_subplot(223)
plt8=fig2.add_subplot(224)
plt9=fig3.add_subplot(221)
plt10=fig3.add_subplot(222)
plt11=fig3.add_subplot(223)
plt12=fig3.add_subplot(224)
plt13=fig4.add_subplot(221)
plt14=fig4.add_subplot(222)
plt15=fig4.add_subplot(223)
plt16=fig4.add_subplot(224)


plt1.plot(result[' _hum'])
plt1.set_title('humidity')
plt2.plot(result[' _rain'])
plt2.set_title('rain')
plt3.plot(result[' _pressurem'])
plt3.set_title('pressure')
plt4.plot(result[' _dewptm'])
plt4.set_title('dew')
plt5.plot(result[' _snow'])
plt5.set_title('snow')
plt6.plot(result[' _fog'])
plt6.set_title('fog')
plt7.plot(result[' _hail'])
plt7.set_title('hail')
plt8.plot(result[' _heatindexm'])
plt8.set_title('heatindex')
plt9.plot(result[' _dewptm'])
plt9.set_title('dew')

plt10.plot(result[' _tempm'])
plt10.set_title('temp')
plt11.plot(result[' _thunder'])
plt11.set_title('thunder')
plt12.plot(result[' _tornado'])
plt12.set_title('tornado')

plt13.plot(result[' _wdird'])
plt13.set_title('wdird')
plt14.plot(result[' _wgustm'])
plt14.set_title('wgustm')
plt15.plot(result[' _windchillm'])
plt15.set_title('windchill')
plt16.plot(result[' _wspdm'])
plt16.set_title('wspdm')



fig1.subplots_adjust(hspace=.6,wspace=0.6)
fig2.subplots_adjust(hspace=.6,wspace=0.6)
fig3.subplots_adjust(hspace=.6,wspace=0.6)
fig4.subplots_adjust(hspace=.6,wspace=0.6)


plt.show()
```

```python
#creating piecharts of the data
import matplotlib.pyplot as pt
activities=['hum','thunder','dewptm','temperature','tornado','wdird']
slices=result[[' _hum',' _thunder',' _dewptm',' _tempm',' _tornado',' _wdird']]
slices1=slices.values.tolist()
colors=['r','g','y','m','g','b']

for i in range(len(slices1)):
 pt.pie(slices1[i],labels=activities,colors=colors,explode=(0,0.1,0,0,0,0.1),autopct="%1.1ff%%")
 pt.title("Day by Day Analysis\n", bbox={ 'pad':5})
 print()   
 pt.show()
 print()



```
